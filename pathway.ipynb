```python
import os

def display_directory_structure(startpath, exclude_dirs=['.git', 'node_modules', '__pycache__']):
    """
    Display directory structure in a tree-like format
    """
    for root, dirs, files in os.walk(startpath):
        # Exclude specified directories
        dirs[:] = [d for d in dirs if d not in exclude_dirs]
        
        level = root.replace(startpath, '').count(os.sep)
        indent = '│   ' * level
        print(f'{indent}├── {os.path.basename(root)}/')
        
        subindent = '│   ' * (level + 1)
        for f in files:
            print(f'{subindent}├── {f}')

# Use it in your notebook
current_dir = os.getcwd()  # Gets current working directory
display_directory_structure(current_dir)
```
